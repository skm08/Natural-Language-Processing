# Latent Dirichlet Allocation(LDA)

```python
import pandas as pd
npr = pd.read_csv('/kaggle/input/npr-topic-modeling/npr.csv', sep='\t')
```

```python
npr.head()
```

```python
npr.shape
```

```python
npr['Article'][10]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_df=0.9,min_df=2,stop_words='english')
dtm = cv.fit_transform(npr['Article'])
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=7,random_state=42)
LDA.fit(dtm)
```

```python
len(cv.get_feature_names_out())
```

```python
type(cv.get_feature_names_out())
```

```python
cv.get_feature_names_out()[50000]
```

```python
import random

random_word_id = random.randint(0,54777)

cv.get_feature_names_out()[random_word_id]
```

```python
len(LDA.components_)
```

```python
LDA.components_.shape
```

```python
LDA.components_
```

```python
single_topic = LDA.components_[0]
single_topic.argsort()
```

```python
single_topic.argsort()[-10:]
```

```python
top_twenty_words = single_topic.argsort()[-20:]
```

```python
for index in top_twenty_words:
    print(cv.get_feature_names_out()[index])
```

```python
for i, topic in enumerate(LDA.components_):
    print(f"The top 15 words for topic #{i}")
    print([cv.get_feature_names_out()[index] for index in topic.argsort()[-15:]])
    print('\n')
```

```python
dtm
```

```python
topic_results = LDA.transform(dtm)
```

```python
topic_results.shape
```

```python
topic_results[0]
```

```python
topic_results[0].argmax()
```

```python
npr['Topic'] = topic_results.argmax(axis=1)
```

```python
npr
```

# Non-Negative Matrix Factorization(NMF)

```python
import pandas as pd
npr = pd.read_csv('/kaggle/input/npr-topic-modeling/npr.csv')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')

dtm = tfidf.fit_transform(npr['Article'])
```

```python
dtm
```

```python
from sklearn.decomposition import NMF

nmf_model = NMF(n_components=7, random_state=42)

nmf_model.fit(dtm)
```

```python
tfidf.get_feature_names_out()[2300]
```

```python
nmf_model.components_
```

```python
nmf_model.components_.shape
```

```python
len(nmf_model.components_)
```

```python
for index, topic in enumerate(nmf_model.components_):
    print(f"The Top 15 Words For Topic # {index}")
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results = nmf_model.transform(dtm)

topic_results.argmax(axis=1)
```

```python
npr['Topic'] = topic_results.argmax(axis=1)
```

```python
npr.head()
```

# Assessment with Quora Questions

```python
import pandas as pd
quora = pd.read_csv('/kaggle/input/quora-questions/quora_questions.csv')
```

```python
quora.head()
```

```python
quora['Question'][0]
```

## Preprocessing

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')

dtm = tfidf.fit_transform(quora['Question'])

dtm
```

## Non-negative Matrix Factorization

```python
from sklearn.decomposition import NMF

nmf_model = NMF(n_components=20, random_state=42)

nmf_model.fit(dtm)
```

```python
for index, topic in enumerate(nmf_model.components_):
    print(f"The top 15 words for topic # {index}")
    print([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-15:]])
    print('\n')
```

```python
topic_results = nmf_model.transform(dtm)

quora['Topic'] = topic_results.argmax(axis=1)
```
